```python
import ipywidgets as widgets
from IPython.display import display
import os

# Create a file upload widget
upload_widget = widgets.FileUpload(accept='image/*', multiple=False)

# Display the widget
display(upload_widget)

# Define the function to handle the upload and save the file
def on_upload_change(change):
    uploaded_file = change['new']
    if uploaded_file:
        # Get the uploaded file's content and filename
        file_name = list(uploaded_file.keys())[0]
        file_content = uploaded_file[file_name]['content']

        # Define the file path where the image will be saved
        save_path = os.path.join("/content", file_name)  # Adjust the directory as needed

        # Save the file to the defined location
        with open(save_path, 'wb') as f:
            f.write(file_content)

        # Print the location of the saved image
        print(f"Logo uploaded and saved at: {save_path}")

# Attach the callback function to handle file uploads
upload_widget.observe(on_upload_change, names='value')



```

```python
import qrcode
import zipfile
from google.colab import files
from PIL import Image, ImageDraw, ImageFont

def create_qr_code_with_logo(link, output_file, logo_path):
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=24,
        border=1.5,
    )
    qr.add_data(link)
    qr.make(fit=True)
    qr_img = qr.make_image(fill_color="black", back_color="white").convert("RGB")
    try:
        logo = Image.open(logo_path)
        logo_width = int(qr_img.size[0] // 1.3)
        logo_height = int(logo.size[1] * logo_width // logo.size[0])
        logo = logo.resize((logo_width, logo_height))
        text_padding = 20
        logo_to_text_space = 110
        combined_height = max(qr_img.size[1], logo.size[1] + text_padding + logo_to_text_space + 30)
        combined_width = qr_img.size[0] + logo.size[0] + 20
        combined_img = Image.new("RGB", (combined_width, combined_height), "white")
        combined_img.paste(qr_img, (0, (combined_height - qr_img.size[1]) // 2))
        logo_x = qr_img.size[0] + 20
        logo_y = (combined_height - logo.size[1] - text_padding - logo_to_text_space) // 2
        combined_img.paste(logo, (logo_x, logo_y))
        draw = ImageDraw.Draw(combined_img)
        font = ImageFont.truetype("ArialCE.ttf", 90)

        # Remove the .jpg extension for text in the image
        display_text = output_file.replace('.jpg', '')

        text_bbox = draw.textbbox((0, 0), display_text, font=font)
        text_width = text_bbox[2] - text_bbox[0]
        text_height = text_bbox[3] - text_bbox[1]
        text_x = logo_x + (logo.size[0] - text_width) // 2
        text_y = logo_y + logo.size[1] + logo_to_text_space
        draw.text((text_x, text_y), display_text, fill="black", font=font)
        right_margin = 50  # Define the width of the right margin
        final_width = combined_width + right_margin
        final_img = Image.new("RGB", (final_width, combined_height), "white")
        final_img.paste(combined_img, (0, 0))  # Paste the original image on the left

        # Save the final image with the right margin
        final_img.save(output_file, 'JPEG')
        print(f"QR code with logo and right margin saved as {output_file}")

    except FileNotFoundError:
        print(f"Logo file not found at {logo_path}")

# Inputs for generating multiple QR codes
logo_path = "/content/logoImg.jpg"  # Path to the logo image

# Ask user how many QR codes to generate
num_qr_codes = int(input("Enter the number of QR codes to generate: "))
link = input("Enter the link: ")
base_output_file = input("Enter the base name for the output file (e.g., 'CRB-06-000'): ")

# List to store generated file names
output_files = []

for i in range(1, num_qr_codes + 1):
    print(f"Creating QR code {i} of {num_qr_codes}")
    output_file = f"{base_output_file}{str(i).zfill(3)}.jpg"
    output_files.append(output_file)

    # Create and save the QR code with the logo
    create_qr_code_with_logo(link, output_file, logo_path)

# Create a ZIP file to bundle all QR codes
zip_filename = "qr_codes.zip"
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for file in output_files:
        zipf.write(file)

# Download the ZIP file
files.download(zip_filename)
print(f"All QR codes downloaded as {zip_filename}")
```
